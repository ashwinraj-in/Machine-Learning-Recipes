## **FEATURE ENGINEERING**

```
test = pd.read_csv("Test Data.csv")
```

```
test.head()
```

```
train = df[["clientid_cr", "pubclientid_cr", "ip_cr", "clmbuserid_cr", "imprid_cr", "siteId_cr", "goalid_cr", "cityId_cr", "stateId_cr", "countryDimId_cr", "browserId_cr", "adslotdimid_cr", "conversiontime_cr", "itemcolumbiaid_cr", "ispDimId_cr", "deviceId_cr", "osId_cr"]]
```

```
test = test[["conversion_fraud"]]
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train,test, test_size = 0.20, random_state = 99)
```

```
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

## **LINEAR REGRESSION**

```
model.fit(train, y_train)
model.score(X_test,y_test)
```
